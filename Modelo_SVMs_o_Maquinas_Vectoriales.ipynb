```python
from sklearn import datasets
datos = datasets.load_breast_cancer(as_frame=True)# para poder ver los datos en un DataFrame de pandas directamente

datos.frame
```

```python
datos.frame.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X = datos.data
y = datos.target
X_ent, X_pru, y_ent, y_pru = train_test_split(X, y,test_size=.3)
```

```python
from sklearn import svm, metrics

kernels = ["linear","rbf","sigmoid"]
gammas = [1, .01, .001, .0001, .000001]

for kernel in kernels:
  for gamma in gammas:
    modelo = svm.SVC(kernel=kernel,gamma=gamma)
    modelo.fit(X_ent,y_ent)
    prediciones = modelo.predict(X_pru)
    print(f"kernel: {kernel} gamma: {gamma} Exactitud:",metrics.accuracy_score(y_pru,prediciones))
    print(f"kernel: {kernel} gamma: {gamma} PresiciÃ³n:",metrics.precision_score(y_pru,prediciones))

#haciendo este ejercicio podemos ver con cual de los kernel le va mejor a nuestro modelo y en este caso nos damos cuenta que al modelo que mejor le fue fue al linear hasta con un 0.98


```

```python
kernel = "linear"
modelo = svm.SVC(kernel=kernel)
modelo.fit(X_ent,y_ent)
prediciones = modelo.predict(X_pru)
print(f"kernel: {kernel} Exactitud:",metrics.accuracy_score(y_pru,prediciones))
print(f"kernel: {kernel} PresiciÃ³n:",metrics.precision_score(y_pru,prediciones))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pru,prediciones))
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
pd.DataFrame(confusion_matrix(y_pru,prediciones))
```
